```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from statsmodels.tsa.arima.model import ARIMA

# Load the dataset
file_path = "/content/air_pollution_data.csv"  # Replace with the actual file path
data = pd.read_csv(file_path)

# Filter data for Ahmedabad
ahmedabad_data = data[data['city'] == 'Ahmedabad']

# Convert 'date' column to datetime and set as index
ahmedabad_data['date'] = pd.to_datetime(ahmedabad_data['date'], format='%d-%m-%Y')
ahmedabad_data.set_index('date', inplace=True)

# Select target variable (e.g., AQI)
time_series = ahmedabad_data['aqi']
```

```python
# Plot time series data
plt.figure(figsize=(12, 6))
plt.plot(time_series, label='AQI')
plt.title('Ahmedabad AQI Over Time')
plt.xlabel('Date')
plt.ylabel('AQI')
plt.legend()
plt.show()
```

```python
# Split into train and test sets (e.g., 80% train, 20% test)
train_size = int(len(time_series) * 0.8)
train, test = time_series[:train_size], time_series[train_size:]
```

```python
# Fit ARIMA model (example with order=(1,1,1))
model = ARIMA(train, order=(1, 1, 1))
model_fit = model.fit()

# Print model summary
print(model_fit.summary())
```

```python
# Forecast on test data
forecast = model_fit.forecast(steps=len(test))

# Plot actual vs predicted values
plt.figure(figsize=(12, 6))
plt.plot(train.index, train, label='Train')
plt.plot(test.index, test, label='Test', color='orange')
plt.plot(test.index, forecast, label='Forecast', color='green')
plt.title('ARIMA Model - AQI Prediction')
plt.xlabel('Date')
plt.ylabel('AQI')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Calculate evaluation metrics
mae = mean_absolute_error(test, forecast)
rmse = np.sqrt(mean_squared_error(test, forecast))

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

```python
# Forecast future values (e.g., next 30 days)
future_forecast = model_fit.forecast(steps=30)

# Plot future forecast
future_dates = pd.date_range(start=test.index[-1], periods=30+1, freq='D')[1:]
plt.figure(figsize=(12, 6))
plt.plot(time_series.index, time_series, label='Historical Data')
plt.plot(future_dates, future_forecast, label='Future Forecast', color='red')
plt.title('Future AQI Forecast for Ahmedabad')
plt.xlabel('Date')
plt.ylabel('AQI')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM

# Load the dataset
file_path = "air_pollution_data.csv"  # Replace with the actual file path
data = pd.read_csv(file_path)

# Filter data for Ahmedabad and process date column
ahmedabad_data = data[data['city'] == 'Ahmedabad']
ahmedabad_data['date'] = pd.to_datetime(ahmedabad_data['date'], format='%d-%m-%Y')
ahmedabad_data.set_index('date', inplace=True)

# Select AQI column for time series
time_series = ahmedabad_data['aqi'].values.reshape(-1, 1)

# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(time_series)

# Create sequences for LSTM (look_back defines the number of previous steps to consider)
def create_sequences(data, look_back=30):
    X, y = [], []
    for i in range(len(data) - look_back):
        X.append(data[i:i + look_back])
        y.append(data[i + look_back])
    return np.array(X), np.array(y)

look_back = 30
X, y = create_sequences(scaled_data, look_back)

# Split into training and testing sets (80% train, 20% test)
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
# Build the LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(look_back, 1)))
model.add(LSTM(units=50))
model.add(Dense(units=1))  # Output layer

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
```

```python
# Make predictions on training and testing sets
train_predictions = model.predict(X_train)
test_predictions = model.predict(X_test)

# Inverse transform predictions and actual values to original scale
train_predictions = scaler.inverse_transform(train_predictions)
test_predictions = scaler.inverse_transform(test_predictions)
y_train_actual = scaler.inverse_transform(y_train)
y_test_actual = scaler.inverse_transform(y_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Calculate evaluation metrics
train_rmse = np.sqrt(mean_squared_error(y_train_actual, train_predictions))
test_rmse = np.sqrt(mean_squared_error(y_test_actual, test_predictions))
train_mae = mean_absolute_error(y_train_actual, train_predictions)
test_mae = mean_absolute_error(y_test_actual, test_predictions)

print(f"Train RMSE: {train_rmse}, Test RMSE: {test_rmse}")
print(f"Train MAE: {train_mae}, Test MAE: {test_mae}")
```

```python
import matplotlib.pyplot as plt

# Plot actual vs predicted values for test set
plt.figure(figsize=(12, 6))
plt.plot(range(len(y_test_actual)), y_test_actual.flatten(), label="Actual AQI")
plt.plot(range(len(test_predictions)), test_predictions.flatten(), label="Predicted AQI", color='orange')
plt.title("Actual vs Predicted AQI (Test Set)")
plt.xlabel("Time")
plt.ylabel("AQI")
plt.legend()
plt.show()
```

```python
# Forecast future values (e.g., next 30 days)
last_sequence = scaled_data[-look_back:]  # Take the last `look_back` values from the dataset
future_forecast = []

for _ in range(30):  # Predict next 30 days
    prediction = model.predict(last_sequence[np.newaxis, :, :])
    future_forecast.append(prediction[0][0])
    last_sequence = np.append(last_sequence[1:], prediction).reshape(-1, 1)

# Inverse transform future forecast to original scale
future_forecast = scaler.inverse_transform(np.array(future_forecast).reshape(-1, 1))

# Plot future forecast
future_dates = pd.date_range(start=ahmedabad_data.index[-1], periods=31)[1:]
plt.figure(figsize=(12, 6))
plt.plot(future_dates, future_forecast.flatten(), label="Future Forecast", color='red')
plt.title("Future AQI Forecast for Ahmedabad")
plt.xlabel("Date")
plt.ylabel("AQI")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional, Attention
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau
from sklearn.metrics import mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt

# 1. Data Loading and Preprocessing
file_path = "/content/air_pollution_data.csv"
data = pd.read_csv(file_path)
ahmedabad_data = data[data['city'] == 'Ahmedabad'].copy()  # Using .copy()
ahmedabad_data['date'] = pd.to_datetime(ahmedabad_data['date'], format='%d-%m-%Y')
ahmedabad_data.set_index('date', inplace=True)

# Features: aqi, co, no2, pm2_5, pm10
features = ['aqi', 'co', 'no2', 'pm2_5', 'pm10']
# Impute missing values using median
for feature in features:
    ahmedabad_data[feature] = ahmedabad_data[feature].fillna(ahmedabad_data[feature].median())

time_series = ahmedabad_data[features].values

# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(time_series)

# Create sequences
def create_sequences(data, look_back=30):
    X, y = [], []
    for i in range(len(data) - look_back):
        X.append(data[i:i + look_back])
        y.append(data[i + look_back, 0])  # Target is AQI
    return np.array(X), np.array(y)

look_back = 30
X, y = create_sequences(scaled_data, look_back)

# Train/Test Split
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# 2. LSTM Model Building
model = Sequential()
model.add(Bidirectional(LSTM(128, return_sequences=True, input_shape=(look_back, len(features)))))
model.add(Dropout(0.3))
model.add(Bidirectional(LSTM(128)))
model.add(Dropout(0.3))
model.add(Dense(1))

# Optimizer and Learning Rate Scheduler
optimizer = Adam(learning_rate=0.001)
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, verbose=1)

model.compile(optimizer=optimizer, loss='mse')

# 3. Model Training
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test), callbacks=[lr_scheduler], verbose=1)

# 4. Predictions
train_predictions = model.predict(X_train)
test_predictions = model.predict(X_test)

# Inverse Transform
train_predictions = scaler.inverse_transform(np.concatenate((train_predictions, np.zeros((len(train_predictions), len(features) - 1))), axis=1))[:, 0]
test_predictions = scaler.inverse_transform(np.concatenate((test_predictions, np.zeros((len(test_predictions), len(features) - 1))), axis=1))[:, 0]
y_train_actual = scaler.inverse_transform(np.concatenate((y_train.reshape(-1, 1), np.zeros((len(y_train), len(features) - 1))), axis=1))[:, 0]
y_test_actual = scaler.inverse_transform(np.concatenate((y_test.reshape(-1, 1), np.zeros((len(y_test), len(features) - 1))), axis=1))[:, 0]

# 5. Evaluation
train_rmse = np.sqrt(mean_squared_error(y_train_actual, train_predictions))
test_rmse = np.sqrt(mean_squared_error(y_test_actual, test_predictions))
train_mae = mean_absolute_error(y_train_actual, train_predictions)
test_mae = mean_absolute_error(y_test_actual, test_predictions)

print(f"Train RMSE: {train_rmse}, Test RMSE: {test_rmse}")
print(f"Train MAE: {train_mae}, Test MAE: {test_mae}")

# 6. Visualization
plt.figure(figsize=(14, 6))
plt.plot(ahmedabad_data.index[look_back:train_size+look_back], y_train_actual, label='Train Actual', color='blue')
plt.plot(ahmedabad_data.index[train_size+look_back:], y_test_actual, label='Test Actual', color='green')
plt.plot(ahmedabad_data.index[look_back:train_size+look_back], train_predictions, label='Train Predicted', color='red', alpha=0.7)
plt.plot(ahmedabad_data.index[train_size+look_back:], test_predictions, label='Test Predicted', color='orange', alpha=0.7)
plt.xlabel('Date')
plt.ylabel('AQI')
plt.title('AQI Prediction with LSTM')
plt.legend()
plt.show()
```
